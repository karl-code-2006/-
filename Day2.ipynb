```python
#将十位老师分配到三个不同的办公室
import random

teachers=['a','b','c','d','e','f','g','h','i','j']
offices=[[],[],[]]

for t in teachers:
    idx=random.randint(0,2)
    offices[idx].append(t)
for i,room in enumerate(offices):
    print(f"办公室{i+1}的人数有{len(room)},老师有:{room}")
```
