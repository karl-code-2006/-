```python
#根据前两个字母判断用户输入的星期
week_dict = {
    'M' : 'Monday:星期一',
    'T' : {'u':'Tuesday:星期二','h':'Thursday:星期四'},
    'W' : 'Wednesday:星期三',
    'F' : 'Friday:星期五',
    'S' : {'a':'Saturday:星期六','u':'Sunday:星期日'}
}
#字典键值对中一定要带逗号

first = input("请输入第一个字母:").upper()
if first in week_dict:
    if  isinstance(week_dict[first],dict):
        #isinstance,判断前面那个元素是否是后面那个类型
        second = input("请输入第二个字母:").lower()
        print(week_dict[first].get(second,"输入错误"))
        #get()函数访问键值对,使用更安全
    else:
        print(week_dict[first])
else:
    print("输入错误")
```

```python
#简易计算器
def calculator(a,b,op):
    if op =='+':return a+b
    elif op == '-':return a-b
    elif op == '*':return a*b
    elif op == '/': 
        return a/b if b != 0 else  "除数不能为0"
    else:return "无定义运算符"
print(calculator(10,0,'/'))
```
