```python
class User:
    def __init__(self,username,password):
        self.username = username
        self.password = password

    def reset_password(self,new_password):
        self.password = new_password
        print(f"用户{self.username}的密码已经修改成功")
u = User("liuao","12345")
print(u.password)
u.reset_password("12347")
print(u.password)
```

```python
class Amphibian:
    def breath(self):
        print("幼年用鳃呼吸,成年用肺呼吸")

class Frog(Amphibian):
    def jump(self):
        print("我会跳跃")

    def breath(self):
        super().breath()
        print("我还能用皮肤呼吸")

f = Frog()
f.breath()
f.jump()
```
