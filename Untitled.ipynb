```python
prices = [199,49,88,299,150,66,30,450,120,200]
min_1=float(input("请输入最小值(元):"))
max_1=float(input("请输入最大值(元)"))
#筛选并排序
price_list=[p for p in prices if min_1 <= p <= max_1]
price_list.sort()
print(f"区间内的商品价格(升序):{price_list}")
```

```python
friends=[]
while True:
    print("\n1.添加好友 2.删除好友 3.修改备注 4.展示好友 0.退出")
    operation=input("请选择操作:")
    if operation == '1':
        friends.append(input("请输入好友姓名:"))
    elif operation == '2':
        r=input("请输入好友姓名:")
        if r in friends:friends.remove(r)
    elif operation == '3':
        r=input("请输入好友需要修改的姓名:")
        if r in friends:old_name=friends.index(r)
        r=input("请输入好友的备注姓名:")
        friends[old_name]=r
    elif operation == '4':
        print(f"好友的列表是:friends")
    elif operation == '0':
        break
```
