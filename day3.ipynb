```python
#阿拉伯数字与汉字数字的转换
chinese_nums = ('零','壹','贰','叁','肆','伍','陆','柒','捌','玖')
num_str = input("请输入阿拉伯数字:")
for char in num_str:
    if char.isdigit():
        print(f"中文大写形式是:{chinese_nums[int(char)]}")
```

```python
#输入首字母输出对应的星期单词
week_dict = {
    'M':'Monday',
    'T':{'u':'Tuesday','h':'Thursday'},
    'W':'Wednesday',
    'F':'Friday',
    'S':{'a':'Saturday','u':'Sunday'}
}

first = input("请输入第一个字母:").upper()
if first in week_dict:
    if isinstance(week_dict[first],dict):
        second = input("请输入第二个字母:").lower()
        print(week_dict[first].get(second,"输入有误"))
    else:
        print(week_dict[first])
else:
    print("输入有误")
```
