```python
#学生信息管理系统
students=[]


def add_stu(name,age):
    global students
    students.append({'name':name,'age':age})
    #字典的添加函数是add(),列表的添加函数是append()
def show_stu():
    for s in students:
        print(f"姓名:{s['name']},年龄:{s['age']}")
        #这种输出发布方式值得记忆
add_stu("张三",19)
#这种需要用户输入""的方式值得改进
show_stu()
    
```

```python
class Account:
    def __init__(self, name, balance=0):
        self.name = name
        self.__balance = balance

    def get_balance(self):
        """获取当前余额（通过公有方法访问私有属性）"""
        return self.__balance

    def deposit(self, amount):
        """存款，amount必须为正数"""
        if amount <= 0:
            print("存款金额必须大于0")
            return
        self.__balance += amount
        print(f"您存款{amount}元，目前余额为：{self.__balance}元")

    def withdraw(self, amount):
        """取款，要求余额充足且金额为正"""
        if amount <= 0:
            print("取款金额必须大于0")
            return
        if amount > self.__balance:
            print("对不起，余额不足")
            return
        self.__balance -= amount
        print(f"您取出{amount}元，目前余额为：{self.__balance}元")

    def show_info(self):
        """显示账号信息"""
        print(f"您的账号信息为：姓名：{self.name}  余额：{self.get_balance()}元")


def main():
    account = None  # 初始无账号
    while True:
        print("\n===== 银行管理系统 =====")
        print("1. 注册账号")
        print("2. 存款")
        print("3. 取款")
        print("4. 查看账号信息")
        print("0. 退出系统")
        choice = input("请输入操作编号（0-4）：")

        if choice == '0':
            print("感谢使用，再见！")
            break

        elif choice == '1':
            name = input("请输入您的姓名：").strip()
            if not name:
                print("姓名不能为空")
                continue
            try:
                balance = float(input("请输入您要存入的初始余额："))
                if balance < 0:
                    print("初始余额不能为负数")
                    continue
            except ValueError:
                print("余额输入无效，请输入数字")
                continue
            account = Account(name, balance)
            print("注册成功！")

        elif choice == '2':
            if account is None:
                print("请先注册账号（选择1）")
                continue
            try:
                amount = float(input("请输入存款金额："))
            except ValueError:
                print("金额输入无效，请输入数字")
                continue
            account.deposit(amount)

        elif choice == '3':
            if account is None:
                print("请先注册账号（选择1）")
                continue
            try:
                amount = float(input("请输入取款金额："))
            except ValueError:
                print("金额输入无效，请输入数字")
                continue
            account.withdraw(amount)

        elif choice == '4':
            if account is None:
                print("请先注册账号（选择1）")
                continue
            account.show_info()

        else:
            print("无效的选择，请输入0-4之间的数字")


if __name__ == "__main__":
    main()
```
